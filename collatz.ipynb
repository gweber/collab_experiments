Collatz period

```
%matplotlib inline
import matplotlib.pyplot as plt
```

```
def rule(i):
    # odd
    if i % 2:
        return 3*i+1
    else:
        return i/2

results = {}
for i in range(100000):
    res = i
    steps = 0
    while res > 1:
        steps -= -1
        res = rule(res)
    else:
        results[i] = steps
```

```
plt.subplots(figsize = (400,20))
plt.plot(range(len(results)), list(results.values()))
plt.show()
```
